```python
import requests
from bs4 import BeautifulSoup
from bs4 import Comment
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/revenue.htm"
data  = requests.get(url).text
```

```python
soup = BeautifulSoup(data, 'html.parser')
```

```python
tables=pd.read_html(url)
```

```python
df=tables[1]
```

```python
df.tail()
```

```python
df2=tables[0]
```

```python
df.columns=['Date','Tesla Quarter Revenue (Millions of US $)']
df
```

```python
df.dtypes
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dropna()
```

```python
df['Tesla Quarter Revenue (Millions of US $)'] = df['Tesla Quarter Revenue (Millions of US $)'].str.replace('$', '')
df['Tesla Quarter Revenue (Millions of US $)'] = df['Tesla Quarter Revenue (Millions of US $)'].str.replace(',', '')
df['Tesla Quarter Revenue (Millions of US $)'] = df['Tesla Quarter Revenue (Millions of US $)'].astype(int)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fig, axes = plt.subplots(2, 1, figsize=(18, 10))

fig.suptitle('Dashboard')

sns.lineplot(ax=axes[0], data=Tesla_Share_Price, x="Date", y="Open")
sns.lineplot(ax=axes[1], data=df, x='Date', y='Tesla Quarter Revenue (Millions of US $)')
fig.tight_layout()
```
