```python
url1 = "https://companiesmarketcap.com/gamestop/revenue/"
datags = requests.get(url1)
```

```python
soup1 = BeautifulSoup(datags.text,'html.parser')
```

```python
soup1
```

```python
tablegm=soup1.find("table",class_="table")
```

```python
headers=tablegm.find_all("th")
Column_names=[c.string for c in headers]
Column_names
```

```python
print('Classes of each table:')
for table in soup1.find_all('table'):
    print(table.get('class'))
```

```python
tablesgm = soup1.find_all('table')
```

```python
tablegm= soup1.find('table', class_='table')
```

```python
table_rows=tablegm.find('tbody').find_all('tr')
```

```python
l=[]
for tr in table_rows:
    td=tr.find_all('td')
    row=[str(tr.string).strip() for tr in td]
    l.append(row)
print(l)
```

```python
dfgm=pd.DataFrame(l,columns=Column_names)
dfgm.drop('Year',axis='columns', inplace=True)
dfgm['Year']=range(2024,2000,-1)
dfgm
```

```python
fig, axes = plt.subplots(1, 2, figsize=(18, 10))

fig.suptitle('Dashboard')

sns.lineplot(ax=axes[0], data=Gamestop_Share_Price, x="Date", y="Open")
sns.lineplot(ax=axes[1], data=dfgm, x='Year', y='Revenue')
fig.tight_layout()
```
